```python
import Pcb_Car 
import time
import threading

# 初始化舵机控制
car = Pcb_Car.Pcb_Car()

# 定义舵机旋转函数，指定舵机号和旋转范围
def rotate_servo_smoothly(servo_id, start_angle, end_angle, step_angle=1, delay=0.05):
    if start_angle < end_angle:
        # 从小角度旋转到大角度
        for angle in range(start_angle, end_angle + 1, step_angle):
            car.Ctrl_Servo(servo_id, angle)
            time.sleep(delay)
    else:
        # 从大角度旋转到小角度
        for angle in range(start_angle, end_angle - 1, -step_angle):
            car.Ctrl_Servo(servo_id, angle)
            time.sleep(delay)

# 定义一个缓慢回归到90度位置的函数
def return_to_initial_position(servo_id, current_angle, initial_angle=90, step_angle=1, delay=0.05):
    # 如果当前角度不在初始位置，则缓慢回到初始位置
    rotate_servo_smoothly(servo_id, current_angle, initial_angle, step_angle, delay)


def reset_initial():
     # 创建线程缓慢匀速地将两个舵机都回归到90度
    servo1_reset_thread = threading.Thread(target=return_to_initial_position, args=(1, 0, 90, 2, 0.05))
    servo2_reset_thread = threading.Thread(target=return_to_initial_position, args=(2, 0, 90, 2, 0.05))

    # 启动复位线程
    servo1_reset_thread.start()
    servo2_reset_thread.start()

    # 等待复位线程执行完毕
    servo1_reset_thread.join()
    servo2_reset_thread.join()
    
    
# 控制两个舵机分别在不同的线程中执行旋转
def demo_two_servos_with_threads():
    
    # 创建线程来控制舵机1的旋转
    servo1_thread_start = threading.Thread(target=rotate_servo_smoothly, args=(1, 0, 180, 5, 0.1))
    # 创建线程来控制舵机2的旋转
    servo2_thread_start = threading.Thread(target=rotate_servo_smoothly, args=(2, 0, 180, 5, 0.1))

    # 启动线程
    servo1_thread_start.start()
    servo2_thread_start.start()

    # 等待两个线程执行完毕
    servo1_thread_start.join()
    servo2_thread_start.join()

    
    # 创建线程来控制舵机1的旋转
    servo1_thread = threading.Thread(target=rotate_servo_smoothly, args=(1, 180, 0, 5, 0.1))
    # 创建线程来控制舵机2的旋转
    servo2_thread = threading.Thread(target=rotate_servo_smoothly, args=(2, 180, 0, 5, 0.1))

    # 启动线程
    servo1_thread.start()
    servo2_thread.start()

    # 等待两个线程执行完毕
    servo1_thread.join()
    servo2_thread.join()

         # 创建线程缓慢匀速地将两个舵机都回归到90度
    servo1_reset_thread = threading.Thread(target=return_to_initial_position, args=(1, 0, 90, 2, 0.05))
    servo2_reset_thread = threading.Thread(target=return_to_initial_position, args=(2, 0, 90, 2, 0.05))

    # 启动复位线程
    servo1_reset_thread.start()
    servo2_reset_thread.start()

    # 等待复位线程执行完毕
    servo1_reset_thread.join()
    servo2_reset_thread.join()

# 执行演示
demo_two_servos_with_threads()
```

```python
import Pcb_Car 
import time
import threading

# 初始化舵机控制
car = Pcb_Car.Pcb_Car()

# 定义舵机旋转函数，指定舵机号和旋转范围
def rotate_servo_smoothly(servo_id, start_angle, end_angle, step_angle=1, delay=0.1):  # 增加延迟
    if start_angle < end_angle:
        for angle in range(start_angle, end_angle + 1, step_angle):
            car.Ctrl_Servo(servo_id, angle)
            time.sleep(delay)
    else:
        for angle in range(start_angle, end_angle - 1, -step_angle):
            car.Ctrl_Servo(servo_id, angle)
            time.sleep(delay)
    
    # 缓慢匀速回到90度位置
    current_angle = end_angle if end_angle <= 180 else 180
    return_to_initial(servo_id, current_angle, 90, step_angle, delay)

# 定义复位函数
def return_to_initial(servo_id, current_angle, target_angle, step_angle=1, delay=0.1):  # 增加延迟
    if current_angle < target_angle:
        for angle in range(current_angle, target_angle + 1, step_angle):
            car.Ctrl_Servo(servo_id, angle)
            time.sleep(delay)
    else:
        for angle in range(current_angle, target_angle - 1, -step_angle):
            car.Ctrl_Servo(servo_id, angle)
            time.sleep(delay)

# 控制两个舵机分别在不同的线程中执行旋转
def demo_two_servos_with_threads():
    servo1_thread = threading.Thread(target=rotate_servo_smoothly, args=(1, 0, 180, 1, 0.1))  # 减少步进角度
    servo2_thread = threading.Thread(target=rotate_servo_smoothly, args=(2, 0, 180, 1, 0.1))  # 减少步进角度

    servo1_thread.start()
    servo2_thread.start()

    servo1_thread.join()
    servo2_thread.join()

# 执行演示
demo_two_servos_with_threads()
```
